```python
metrics, filtered_data = neuradock.eeg_quality_check(bluetooth_data, fs=250)
```

```python
# 4. 数据清洗
# 使用上一步返回的 filtered_data 和 metrics
thresh = [10, 20, 1]  ##50Hz ,EMG, artificial points
clean_data, mask, info = neuradock.clean_eeg_data(filtered_data, metrics, thresh)
```
